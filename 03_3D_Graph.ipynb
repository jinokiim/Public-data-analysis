### 데이터 불러오기

```python
import pandas as pd

store_2019 = pd.read_csv('store_2019.csv',encoding='utf-8')
store_2019.head(2)

store_2020 = pd.read_csv('store_2020.csv',encoding='utf-8')
store_2020.head(2)
```

### 컬럼 이름 바꾸기
컬럼 이름이 한글이여서 이후에 간단하게 사용하기 위해 영어 이름으로 변경

```python
rename = pd.read_csv('rename.csv', encoding='utf-8')

store_2019.columns = list(rename['바꾼이름'])
store_2020.columns = list(rename['바꾼이름'])

store_2019.head(2)
```

### 3D 그래프 준비

```python
x_axis = ['0-6시','6-11시','11-14시','14-17시','17-21시','21-24시']
y_axis = store_2020.groupby('service_type')['Q_count'].sum().sort_values(ascending=False)[:10].index
```

```python
store_2020.groupby('service_type')['Q_count'].sum().sort_values(ascending=False)[:10]
```

```python
df = store_2020[store_2020['service_type'].isin(y_axis)].groupby('service_type').sum().iloc[:,60:66]
df
```

[x좌표, y좌표, 매출건수] 형태로 표현

```python
import numpy as np
data=[]
for i in range(10):
    for j in range(6):
        data.append([i,j,np.array(df.values)[i][j]])
data
```

### 그래프 그리기

```python
from pyecharts import Bar3D

bar3d = Bar3D("2020년 Top10 업종,시간대별 매출건수 3D Graph", width=1200, height=600)

bar3d.add(
    "",
    ['0-6시','6-11시','11-14시','14-17시','17-21시','21-24시'],
    df.index,
    [[d[1], d[0], d[2]] for d in data],
    is_visualmap=True,
    visual_range=[0, 180000000],
    grid3d_width=100,
    grid3d_depth=150,
)
bar3d.width=700
bar3d.height=500

bar3d
```
